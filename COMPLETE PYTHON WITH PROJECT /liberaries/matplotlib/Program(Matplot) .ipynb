# Bar Graph

```python
x=[1,2,3,4]
y=[5,6,7,8]

plt.bar(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
c=['r','g','b','y']
plt.bar(x,y,color=c)
plt.show()
```

# Scatter plot 

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[1,4,5,8,7,3]
plt.title("scatter plot")
plt.xlabel("month")
plt.ylabel("days")
plt.scatter(x,y)

plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[1,4,5,8,7,3]
plt.title("scatter plot")
plt.xlabel("month")
plt.ylabel("days")
c=['red','blue ','red','green','brown','orange']
plt.scatter(x,y)

plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[1,4,5,8,7,3]
plt.title("scatter plot")
plt.xlabel("month")
plt.ylabel("days")
c =['red','blue ','magenta','green','brown','orange']
plt.scatter(x,y , c ='c')

plt.show()
```

```python
# immport Libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname= r'bird.png'

#opening an Image

image=Image.open(fname).convert('L')

#mapping image to grey coLor
plt.imshow(image, cmap='gray')
plt.title("Bird")
plt.xlabel("length")
plt.ylabel("breadth")
plt.grid()
plt.show()
```

# Matplotlib pie-chart

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=['maths','english','science','sst']
c=['red','yellow','magenta','aqua']
plt.pie(x,labels=y,colors=c)
plt.show()
```
