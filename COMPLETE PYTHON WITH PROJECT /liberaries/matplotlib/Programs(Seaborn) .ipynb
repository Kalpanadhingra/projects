# bar graph in seaborn

```python
import seaborn as sns

import matplotlib.pyplot as plt

import pandas as pd
```

```python

df1=sns.load_dataset("penguins")
```

```python
df1
```

```python
df1["island"]
```

```python
sns.barplot(x=df1.island,y=df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df1, hue="sex")
plt.show()
```

# order parameter

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df1, hue="sex", order=order1)
plt.show()
```

# hue_order

```python
order1=["Dream","Torgersen","Biscoe"]
order2=["Female","Male"]
sns.barplot(x="island",y="bill_length_mm",data=df1, hue="sex", order=order1, hue_order=order2)
plt.show()
```

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df1, hue="sex", order=order1, hue_order=["Female","Male"])
plt.show()
```

```python
order1=['Dream','Torgersen','Biscoe']
color1=['pink','red','green']
sns.barplot(x="island",y="bill_length_mm",data=df1,order=order1,color='red', hue="sex",palette='prism')
```

# saturation

```python
sns.barplot(x="island",y='bill_length_mm',data=df1,hue="sex",palette="gist_heat_r",saturation=1.3)
plt.show()
```

```python
sns.barplot(x="island",y='bill_length_mm',data=df1,hue="sex",palette="prism",saturation=0.5)
plt.show()
```

# bins

# color

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],color='red')
plt.show()
```

# kde

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde="True",color='g')
plt.show()
```
