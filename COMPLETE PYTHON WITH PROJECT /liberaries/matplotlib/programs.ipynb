# adding titles

```python
fi= go.Figure(go.Scatter(x=x,y=y))
```

```python
fi.update_layout(title="linegraph",xaxis_title="This is a x axis",yaxis_title="This is y axis")
fi.show()
```

```python
x1,y1=[1,2,3,4,5],[5,6,7,8,9]
x2,y2=[11,2,3,4,5],[3,4,2,1,12]
x3,y3=[2,4,5,8,1],[4,6,8,2,0]
```

```python
fig=go.Figure()
```

```python
fig.add_trace(go.Scatter(x=x1,y=y1))
fig.add_trace(go.Scatter(x=x2,y=y2))
fig.add_trace(go.Scatter(x=x3,y=y3))
```

fig=go.Figure()

```python
fig.add_trace(go.Scatter(x=x1,y=y1,name="line1",mode="lines"))
fig.add_trace(go.Scatter(x=x2,y=y2,name="line2",mode="markers"))
fig.add_trace(go.Scatter(x=x3,y=y3,name="line3",mode="lines+markers"))
```

# bubblecharts

```python
import plotly.graph_objects as go
```

```python
x=[12,4,3]
y=[3,4,5,6]
fig=go.Figure(go.Scatter(x=x,y=y, mode="markers"))
fig.show()
```

```python
x=[12,4,3]
y=[3,4,5,6]
fig=go.Figure(go.Scatter(x=x,y=y, mode="lines+markers"))
fig.show()
```

```python
x=[12,4,3,1]
y=[3,4,5,6]
fig=go.Figure(go.Scatter(x=x,y=y, mode="lines+markers",marker_size=[20,30,40,50]))
fig.show()
```

```python
x=[12,4,3,1]
y=[3,4,5,6]
fig=go.Figure(go.Scatter(x=x,y=y, mode="lines+markers",marker_size=[20,30,40,50],text=["productA","productB","productC","productD"]))
fig.show()
```
